```python
import requests
from requests.auth import HTTPBasicAuth
url ="http://url/"
username="username"
password= 'password'

print("Securing Connection ...")
r=requests.get(url + 'api/resources',auth=HTTPBasicAuth(username, password))
if r.ok:
    print(r.status_code)
    print(r.headers['content-type'])
else:
    print("Error")
```

```python
dh=r.json() # Get the content of the api in JSON format
#Create a new list of dict by selecting only displayName as key and href as value
dhis=[]
for v in dh['resources']:
    a={v['displayName']:v['href']}
    a=dict(a)
    dhis.append(a)
        
```

```python
#list(map(lambda d: d['Data Elements'], dhis))
# Get API Url of an endpoint ie displayName
for d in dhis:
    for i,v in d.items():
        if i == 'Data Sets':
            print(v)
```

```python
#All the endpoints
endpts=[ i for d in dhis for i,v in d.items()]
endpts
```

```python
# [Updated ]Get API Url of an endpoint ie displayName
req_url=[ v for d in dhis for i,v in d.items() if i == 'Data Elements']
req_url=str(req_url[-1])
```

```python
#Creating session and fetching data for one API endpoint == Data Sets
s = requests.Session()
print("Securing Connection ...")
d=s.get(req_url,auth=HTTPBasicAuth(username, password), )
if d.ok:
    print(d.status_code)
    print(d.headers['content-type'])
```

```python
#Creating session and fetching data for all API endpoint
for endpt in endpts:
    s = requests.Session()
    q_url=[ v for d in dhis for i,v in d.items() if i == endpt]
    q_url=str(q_url[-1])
    dat=s.get(q_url,auth=HTTPBasicAuth(username, password))
    if dat.ok:
        print()
        print("<================  %s  ===============>" %endpt)
        print(dat.headers['content-type'])
        data=dat.json()
        #low = lambda s: s[:1].lower() + s[1:] if s else '' #Converts first letter to lowercase
        pd.DataFrame(data[list(data.keys())[-1]])
```
